```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/career_compass_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['aptitude_score','interest_alignment_score','personality_match_score','skill_readiness_score','academic_strength_score']
signal_avgs = df[signal_cols].mean()
colors = ['#8b5cf6','#3b82f6','#10b981','#ef4444','#f59e0b']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('Counselling FYI - Average Career Guidance Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Career Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=70, color='green', linestyle='--', alpha=0.5, label='Healthy threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
interest_counts = df['career_interest'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(interest_counts.index, interest_counts.values, color=colors[:len(interest_counts)])
plt.title('Counselling FYI - Career Interest Area Distribution', fontsize=14, fontweight='bold')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
